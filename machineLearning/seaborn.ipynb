```python
import seaborn as sns
```

```python
#basic ploting with seborn

tips=sns.load_dataset('tips')
tips
```

```python
#create scatter plot-->how data is distributede
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("scatter plot using seaborn")
```

```python
##line plot

sns.lineplot(x='size',y='total_bill',data=tips,color='red')
plt.title("line plot")
plt.show()
```

```python
#bar plot--->categorical plot

sns.barplot(x='day', y='total_bill', data=tips, color='skyblue')
plt.title("bar plot for categorical values")
plt.show()
```

```python
#box plot-->IDENTIFY THE OUTLIERS

sns.boxplot(x='day',y='total_bill',data=tips,color='pink')
plt.title('box plot')
plt.show()
```

```python
#violin plot--->Understand data distribution (Is it skewed? Bimodal? Symmetrical?)

sns.violinplot(x='day', y='total_bill', data=tips,color='brown')
plt.title("violin plot")
plt.show()
```

```python
##histrogram

sns.histplot(tips['total_bill'], bins=10, kde=True)#kernal distributed it show the shape
```

```python
#kde plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
#relationship between each and every feature
#pairplot

sns.pairplot(tips)
```

```python
#heatmap--->we only use the numerical vale

corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd

saled_df=pd.read_csv('sales.csv')
saled_df.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Category',y='Total',data=saled_df,estimator=sum)
plt.title('sales data')
plt.xlabel('Category')
plt.ylabel('Total')
plt.show()
```

```python
sns.barplot(x='Region', y='Quantity', data=saled_df, estimator=sum)
plt.title('bar plot')
plt.xlabel('Region')
plt.ylabel('Quantity')
plt.show()
```
