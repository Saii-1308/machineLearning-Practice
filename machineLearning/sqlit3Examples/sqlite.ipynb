```python
#sql and sqlite

import sqlite3
```

```python
#connect to an SQLite database

connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
#create table

cursor.execute('''
Create Table if Not Exists employee(
               id Integer Primary key,
               name Text Not Null,
               age Integer,
               department text
            )
''')

#commit thechanges
connection.commit()
```

```python
#working with sales data
connection=sqlite3.connect('saless.db')
connection

cursor=connection.cursor()

cursor.execute('''
create table if not exists sales(
               id integer primary key,
               date varchar(100) not null,
               product varchar(100),
               sales integer,
               region varchar(100))
''')

saless_data=[
    ('2023-01-01','product1',100,'North'),
    ('2023-01-02','product2',100,'south'),
    ('2023-01-03','product1',100,'west'),
    ('2023-01-04','product3',100,'east'),
    ('2023-01-05','product2',100,'North'),
]
cursor.executemany('''
insert into sales(date,product,sales,region)
                   values(?,?,?,?)
''',saless_data)

connection.commit()
```

```python
cursor.execute('select * from sales')
rows=cursor.fetchall()

for row in rows :
    print(row)
```

```python
#colse connection
connection.close()
```

```python
import sqlite3
connection=sqlite3.connect('student.db')
connection
```

```python
cursor=connection.cursor()
cursor
```

```python
cursor.execute('''
create table if not exists student(
               id integer primary key,
               name varchar(100) not null,
               address varchar(100),
               age integer,
               department integer)
        ''')
connection.commit()
```

```python
student_data=[
    ('sai','pimperi',20,1),
    ('aditi','nigdi',21,2),
    ('damini','chinchwad',20,1),
    ('sayali','chakan',23,3),
    ('pal','pune',22,2),
    ('ishwari','katraj',19,3),
    ('nikita','kothrud',21,1),
]

cursor.executemany('''
insert into student (name,address,age,department)
                   values(?,?,?,?)
''',student_data)

connection.commit()
```

```python
cursor.execute('select * from student')
rows=cursor.fetchall()

for row in rows :
    print(row)
```

```python
cursor.execute('''
update student set age=21 
               where id=1''')
connection.commit()
```

```python
cursor.execute('select * from student')
rows=cursor.fetchall()

for row in rows :
    print(row)
```

```python
cursor.execute('''
delete from student where id=7''')

connection.commit()
```

```python
cursor.execute('select * from student')
rows=cursor.fetchall()

for row in rows :
    print(row)
```

```python
# Add a new column 'marks' to the student table
cursor.execute('ALTER TABLE student ADD COLUMN marks INTEGER')
connection.commit()

# List of marks to assign to each student (must match the number of students)
marks_list = [80, 90, 69, 89, 78, 90]

# Update each student's marks by their id
for idx, marks in enumerate(marks_list, start=1):
	cursor.execute('UPDATE student SET marks = ? WHERE id = ?', (marks, idx))
connection.commit()
```

```python
cursor.execute('select * from student')
rows=cursor.fetchall()

for row in rows :
    print(row)
```

```python
cursor.execute('''
select * from student where department=1''')
rows=cursor.fetchall()

for row in rows :
    print(row)
```

```python
s = "s"  # Set the prefix you want to search for

query = "select * from student where name like ?"

cursor.execute(query, (f"{s}%",))
result = cursor.fetchall()
rows = result
```
