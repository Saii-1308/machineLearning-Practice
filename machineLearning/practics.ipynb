```python
#class object

class people:

    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def printData(self):
        print("name of person: ", self.name)
        print("age of person: ",self.age)

p1=people("sai",20)
p1.printData()
```

```python
class student(people):
    def __init__(self,id,name,age,address):
        super().__init__(name,age)
        self.id=id
        self.address=address
    
    def printData(self):
        super().printData()
        print("id of student: ",self.id)
        print("address of student: ",self.address)

s1=student(1,"sai",20,"pimperi")
s1.printData()
```

```python
def negativeNum(nums):
    try:
       for num in nums:
            if num<0:
                raise ValueError("negative number are present in the list !!")
           # if num is not int:
              #  raise ValueError
    except negativeNum as n:
        print("negative numberr error are raise and catch in except block!!")
        print(n)
    except ValueError as v:
        print("value error are raise")
        print(v)
    else:
        for num in nums:
            print(num)
    finally:
        print("this finally block is always run!!")

nums=[1,2,0,-4,-1,34]
negativeNum(nums)
    
```

```python
# Map string ranges to numeric values
wkly_study_hours_map = {
	'< 5': 2.5,
	'5 - 10': 7.5,
	'10 - 15': 12.5,
	'15 - 20': 17.5,
	'20 <': 22.5
}

# Convert the column to numeric using the mapping
df['WklyStudyHours_num'] = df['WklyStudyHours'].map(wkly_study_hours_map)

# Fill missing values with the mean
mean_val = df['WklyStudyHours_num'].mean()
df['new'] = df['WklyStudyHours_num'].fillna(mean_val)
```
