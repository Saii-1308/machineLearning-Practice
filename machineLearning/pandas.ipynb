```python
#merging and joining dataframes
data1=pd.DataFrame({'key':['A','B','C'],'value1':[1,2,3]})
data2=pd.DataFrame({'key':['A','B','C'],'value2':[4,5,6]})
```

```python
data1
```

```python
data2
```

```python
#merge dataframes on key columns
pd.merge(data1,data2,on='key',how='outer')#on='column_name'--->which column are match in 2nd dataframe
```

```python
pd.merge(data1,data2,on='key',how='left')
```
